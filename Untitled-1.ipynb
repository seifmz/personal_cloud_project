```python
# Initialize the EC2 client and resource with AWS credentials and region
clientec2_client = boto3.(
    'ec2',
    aws_access_key_id=aws_access_key_id,
    aws_secret_access_key=aws_secret_access_key,
    aws_session_token=aws_session_token,
    region_name=region_name
)

ec2_resource = boto3.resource(
    'ec2',
    aws_access_key_id=aws_access_key_id,
    aws_secret_access_key=aws_secret_access_key,
    aws_session_token=aws_session_token,
    region_name=region_name
)

# Function to delete EC2 instances
def delete_ec2_instances():
    """
    Delete all EC2 instances in the specified AWS region.

    This function iterates over all running or stopped EC2 instances
    in the AWS region defined in the ec2_resource and terminates them.
    It waits for each instance to be fully terminated before proceeding to the next.
    """
    
    
    instances = ec2_resource.instances.filter(
        Filters=[{'Name': 'instance-state-name', 'Values': ['running', 'stopped']}]
    )
    for instance in instances:
        print(f"Terminating EC2 instance: {instance.id}")
        instance.terminate()
        instance.wait_until_terminated()  # Wait for the instance to terminate

# Function to delete the key pair
def delete_key_pair(key_name='MyKeyPair'):
    """
    Delete a specified EC2 key pair.

    Parameters:
    key_name (str): The name of the key pair to delete.
    """
    print(f"Deleting key pair: {key_name}")
    ec2_client.delete_key_pair(KeyName=key_name)

# Function to delete the security group
def delete_security_group(group_name='MySecurityGroup'):
    """
    Delete a specified EC2 security group.

    Parameters:
    group_name (str): The name of the security group to delete.

    This function attempts to delete the specified security group.
    It catches and prints any exceptions encountered, such as if the group does not exist
    or if there are dependencies preventing its deletion.
    """
    try:
        print(f"Deleting security group: {group_name}")
        ec2_client.delete_security_group(GroupName=group_name)
    except Exception as e:
        print(f"Error deleting security group: {e}")

# Delete resources
delete_ec2_instances()
delete_key_pair()
delete_security_group()
```
